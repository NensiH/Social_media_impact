```python
# Import our dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in cvs as dataframe
suicide_df = pd.read_csv("../Data/socialmedia_mentalhealth.csv")
suicide_df.head()
```

```python
# change column name 'suicide_year' to 'year'
suicide_df.rename(columns={'suicide_year': 'year'}, inplace=True)
suicide_df
```

```python
# Remove year 2016 from dataframe and reset the index
first_filter = suicide_df.year != 2016
suicide_df = suicide_df[first_filter]
suicide_df.reset_index(drop=True, inplace=True)
suicide_df
```

```python
# Create the output file (CSV).
output_data_file = "../Data/cleaned_socialmedia_mentalhealth.csv"
# Export the usage dataframe into a CSV.
suicide_df.to_csv(output_data_file, index = False)
```
