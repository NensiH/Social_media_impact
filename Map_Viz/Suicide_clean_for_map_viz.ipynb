```python
import pandas as pd
import numpy as np
```

```python
suicide_df = pd.read_csv("Kaggle_suicide_data.csv")
suicide_df.head()
```

```python
# drop the columns not needed for map visualization
suicide_df = suicide_df.drop(
    ['HDI for year', 'gdp_per_capita ($)', 'country-year', 'gdp_per_capita ($)', 
     'age', 'population', 'suicides_no', 'year', ' gdp_for_year ($) '], axis=1)
suicide_df.head()
```

```python
# check for null values
suicide_df.info()
```

```python
# calculate per-country mean
suicide_mean = pd.DataFrame(suicide_df['suicides/100k pop'].groupby(suicide_df['country']).mean())
suicide_mean
```

```python
print(suicide_mean['suicides/100k pop'].max())
print(suicide_mean['suicides/100k pop'].min())
print(suicide_mean['suicides/100k pop'].mean())
```

```python
# group by average per-capita suicides on country and generation
country_per_capita_df = pd.DataFrame(suicide_df['suicides/100k pop']
                                     .groupby([suicide_df['country'], suicide_df['generation']])
                                     .mean())
country_per_capita_df
```

```python
country_per_capita_unstacked = country_per_capita_df.unstack(level=1)
country_per_capita_unstacked
```

```python
# export as CSV
output_csv_mean = "suicide_clean_country_mean.csv"
output_csv_unstacked = "suicide_clean_generation_country_mean.csv"

suicide_mean.to_csv(output_csv_mean)
country_per_capita_unstacked.to_csv(output_csv_unstacked)
```
